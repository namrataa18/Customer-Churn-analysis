```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Determine subplot grid size
n_cols = 3
n_rows = -(-len(columns) // n_cols)  # Ceiling division

# Set up the figure
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()  # Flatten in case of single row/column

# Loop through each column and plot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue =df["Churn"])
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=0)

# Hide unused subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#The data visualizations highlight customer churn patterns across various telecom services. Customers lacking support services like Online Security, Tech Support, and Online Backup show higher churn rates. Fiber optic internet users also exhibit more churn compared to DSL users. Overall, limited service engagement is associated with increased customer churn.
